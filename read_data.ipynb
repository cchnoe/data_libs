```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import os 
import matplotlib.pyplot as plt
```

```python
def normalize(data):
    from sklearn import preprocessing
    n = len(data)
    #returns a numpy array
    array = np.array(data['intensity'])
    scaler = preprocessing.MinMaxScaler()
    scaled = scaler.fit_transform(array.reshape(n,-1))
    data['intensity'] = scaled[:,0]
    return data
```

```python
def plot_spectrum(data, title, file):
    plt.figure(figsize=(15,8))
    plt.xlabel("wavelength")
    plt.ylabel("intensity")
    plt.plot(data['wavelength'], data['intensity'])
    plt.title(str(title), 3
          fontdict={'family': 'serif', 
                    'weight': 'bold',
                    'size': 14})
    plt.savefig('data_output/'+ file +'.png')
    plt.show()
    return
```

```python
def str_to_int(data_o):
    data = data_o.copy()
    wavelength = []
    intensity = []
    for i in data['wavelength']:
        wavelength.append(float(i.replace(',','.')))
    for i in data['intensity']:
        intensity.append(float(i.replace(',','.')))
    df_dict = {'wavelength':wavelength, 
               'intensity':intensity}
    return pd.DataFrame(df_dict)
```

```python
path_rpt = "data_input/"
list_rpt = os.listdir(path_rpt)
```

```python
columns = ["wavelength", "intensity"]
data = {}
data_cu = {}
data_au = {}
data_ag = {}
for file in list_rpt:
    if os.path.isfile(os.path.join(path_rpt, file)) and file.startswith('Cu') and file.endswith('txt'):
        df_file = pd.read_csv(path_rpt + file,sep="\t", header = None, names = columns)
        data_cu[file] = df_file
    elif os.path.isfile(os.path.join(path_rpt, file)) and file.startswith('au') and file.endswith('txt'):
        df_file = pd.read_csv(path_rpt + file,sep="\t", header = None, names = columns)
        data_au[file] = df_file
    elif os.path.isfile(os.path.join(path_rpt, file)) and file.startswith('Ag') and file.endswith('txt'):
        df_file = pd.read_csv(path_rpt + file,sep="\t", header = None, names = columns)
        data_ag[file] = df_file       
data = {'cu':data_cu, 'ag':data_ag, 'au':data_au}
```

```python
data_cu = []
data_au = []
data_ag = []
for element in data.keys():
    for file in data[element].keys():
        data[element][file] = str_to_int(data[element][file]) 
        if file.startswith('Cu'):
            data_cu.append(data[element][file]['intensity'])
            mean = np.mean(data_cu,axis=0 )
            dict_cu = {'wavelength':data[element]['Cu_8.txt']['wavelength'], 'intensity':mean}
            df_cu = pd.DataFrame(dict_cu)
        elif file.startswith('au'):
            data_au.append(data[element][file]['intensity'])
            mean = np.mean(data_au,axis=0 )
            dict_au = {'wavelength':data[element]['au_8.txt']['wavelength'], 'intensity':mean}
            df_au = pd.DataFrame(dict_au)
        elif file.startswith('Ag'):
            data_ag.append(data[element][file]['intensity'])
            mean = np.mean(data_ag,axis=0 )
            dict_ag = {'wavelength':data[element]['Ag_8.txt']['wavelength'], 'intensity':mean}
            df_ag = pd.DataFrame(dict_ag)
```
